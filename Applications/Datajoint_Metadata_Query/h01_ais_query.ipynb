# Finding the AIS

```python
import networkx as nx

def find_ais(graph: nx.Graph, source_node: str) -> str:
    """
    Find the node closest (by hop count) to a node with a specific name
    that has the property compartment == 'axon'.
    
    Args:
        graph (nx.Graph): The input NetworkX graph.
        source_name (str): The 'name' value of the source node.
        name_attr (str): The attribute used to identify the source node (default: 'name').
    
    Returns:
        The node ID (or label) of the closest axon node, or None if not found.
    """
    # Step 1: Perform BFS to find shortest paths from source_node
    for target_node in nx.bfs_tree(graph, source=source_node):
        if target_node == source_node:
            continue
        data = graph.nodes[target_node]
        if data.get('compartment') == 'axon':
            return target_node

    return None  # No axon node found
```
